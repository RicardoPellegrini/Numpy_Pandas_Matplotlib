# Principal Component Analysis (PCA)
An√°lise dos Principais Componentes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris['data']
y = iris['target']
```

```python
X.shape, y.shape
```

```python
X = StandardScaler().fit_transform(X)
```

```python
iris['feature_names']
```

```python
pd.DataFrame(X, columns=iris['feature_names']).head()
```

```python
pca = PCA(n_components=2)
```

```python
pc = pca.fit_transform(X)
```

```python
pc.shape
```

```python
pc[:10]
```

```python
plt.scatter(pc[:,0], pc[:, 1])
```

```python
plt.scatter(pc[y==0,0], pc[y==0,1], label='Setosa')
plt.scatter(pc[y==1,0], pc[y==1,1], label='Versicolor')
plt.scatter(pc[y==2,0], pc[y==2,1], label='Virginica')
plt.legend()

plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
# Foi retida quase 96% das informa√ß√µes com a redu√ß√£o
sum(pca.explained_variance_ratio_)
```

# D√≠gitos

```python
digitos = datasets.load_digits()
```

```python
X = digitos['data']
y = digitos['target']
```

```python
X.shape, y.shape
```

```python
pca = PCA(n_components=2)
X_reduzido = pca.fit_transform(X)

plt.figure(figsize=(12,10))
plt.scatter(X_reduzido[:,0], X_reduzido[:,1],
            c=y, edgecolor='black', alpha=0.6, 
            s=80, cmap='nipy_spectral')
```

# TSNE

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(random_state=123)
```

```python
X_tsne = tsne.fit_transform(X)
```

```python
X_tsne.shape
```

```python
plt.figure(figsize=(12,10))
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, 
            edgecolors='black', alpha=0.6, 
            s=80, cmap='nipy_spectral')
```

```python
# alternativa, mantendo integridade de 90%, ai vai calcular o n√∫mero de componentes
pca = PCA(0.90)
```

```python
pca.fit(X)
```

```python
# Precisa ter 21 atributos para 90% de integridade
pca.n_components_
```

```python
pca.n_components
```

```python
plt.figure(figsize=(8,4))
plt.bar(np.arange(1, pca.n_components_+1), 
        pca.explained_variance_ratio_[:pca.n_components_],
       color='lightskyblue')
plt.xlabel('Componentes principais')
plt.ylabel('Varia√ß√£o')
plt.xticks(np.arange(1, pca.n_components_+1))
plt.show()
```
